```python
import pandas as pd
import ast
import joblib
import ast
from url_helper import URLFeatureExtractor
```

```python
df = pd.read_csv("../datasets/Phishing_Email_URLS.csv")
```

```python
df.head()
```

```python
def concatenate_url(parts):
    if not parts:
        return ""
    return f"{parts[0]}://{parts[1]}{parts[2]}"
```

```python
df['Extracted URLs'] = df['Extracted URLs'].apply(ast.literal_eval)
```

```python
df['full_urls'] = df['Extracted URLs'].apply(lambda x: [concatenate_url(url) for url in x])
```

```python
df['num_urls'] = df['full_urls'].apply(len)
```

```python
df['email_length'] = df['Email Text'].apply(len)
```

```python
df.head()
```

```python
df.rename(columns={"Email Text":"email_text", "Email Type": "email_type", "Contains_URL": "contains_url", "Extracted URLs": "urls_parsed"})
```

```python
df.to_csv("../datasets/Phishing_Email_URLS.csv", index=False)
```

```python
with open('../models/best_email_phishing_model.pkl', 'rb') as f:
    email_text_model = joblib.load(f)
```

```python
df['email_text_predictions'] = email_text_model.predict(df['Email Text'])
df.head()
```

```python
with open('../models/best_url_phishing_model.pkl', 'rb') as f:
    url_model = joblib.load(f)
```

```python
def predict_urls(row):
    if row['num_urls'] == 0:
        return 0  # considered safe if no URLs
    else:
        urls = ast.literal_eval(row['full_urls'])
        # Returning 1 for the first positive url
        for url in urls:
            features = URLFeatureExtractor().extract_features(url)
            feature_names, values = zip(*features.items())
            prediction_input = pd.DataFrame([values], columns=feature_names)
            if url_model.predict(prediction_input)[0] == 1:
                return 1
        return 0
```

```python
df['url_predictions'] = df.apply(predict_urls, axis=1)
```

```python
from transformers import pipeline
```

```python
sentiment_pipeline = pipeline('sentiment-analysis')
```

```python
def get_sentiment_label(text):
    max_length = 512
    tokens = sentiment_pipeline.tokenizer(text, max_length=max_length, truncation=True)
    truncated_text = sentiment_pipeline.tokenizer.decode(tokens['input_ids'], skip_special_tokens=True)

    result = sentiment_pipeline(truncated_text)[0]
    return result['label'], result['score']
```

```python
df[['sentiment_label', 'sentiment_score']] = df['Email Text'].apply(lambda x: pd.Series(get_sentiment_label(x)))
```

```python
meta_features = df[['email_text_predictions', 'url_predictions', 'num_urls', 'email_length', 'Email Type']]
```

```python
meta_features = meta_features.rename(columns={"Email Type": "label"})
meta_features["label"] = meta_features["label"].replace({"Safe Email": 0, "Phishing Email": 1})
```
